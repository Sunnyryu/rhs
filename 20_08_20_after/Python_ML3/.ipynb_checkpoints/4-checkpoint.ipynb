```python
#Pregnancies : 임신 횟수
#Glucose : 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도
#BloodPressure : 이완기 혈압 (mm Hg)
#SkinThickness : 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값
#Insulin : 2시간 혈청 인슐린 (mu U / ml)
#BMI : 체질량 지수 (체중kg / 키(m)^2)
#DiabetesPedigreeFunction : 당뇨병 혈통 기능
#Age : 나이
#Outcome : 768개 중에 268개의 결과 클래스 변수(0 또는 1)는 1이고 나머지는 0입니다.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
# 구버전 주비터 노트북에서는 시각화를 보기 위해서 #matplotlib inline이 필요함
```

```python
df["Pregnancies_high"] = df["Pregnancies"] > 6
df[["Pregnancies", "Pregnancies_high"]].head()
```

## CSV File 저장하기

```python
df.to_csv("data/diabetes_feature.csv", index=False)
# index를 false 해주면 unname가 저장되지 않음
```

```python
df.shape[0]
#768
df.shape[1]
# 데이터를 학습세트와 test세트로 나눠줘야함! 
# [0]으로 인덱싱하면 앞자리인 df의 행의 갯수를 출력 / int로 하여야 정수로 떨어지기에..
split_count = int(df.shape[0] * 0.8)
split_count
```

```python
df[df["Insulin_nan"] > 600].shape
```

```python
train = df[:split_count].copy()
# 1~ 614번째까지 넣어줌!
train.shape
```

```python
train[train["Insulin_nan"] < 600].shape
```

```python
train = train[train["Insulin_nan"] < 600]
train.shape
```

```python
test = df[split_count:].copy()
# 614~ 끝까지를 넣어줌!
test.shape
```

```python
feature_names= train.columns.tolist()
#여러개이기 때문에 리스트로 변경하여 저장해줌!
feature_names.remove("Pregnancies")
feature_names.remove("Outcome")
feature_names.remove("Age_low")
feature_names.remove("Age_middle")
feature_names.remove("Age_high")
feature_names.remove("Insulin")
feature_names.remove("Insulin_log")
feature_names
```

```python
label_name = 'Outcome'
label_name
```

```python
# 학습세트 만들기
X_train = train[feature_names]
print(X_train.shape)
# outcome을 뺸 column이기에 8로 나와야함!
X_train.head()
```

```python
y_train = train[label_name]
print(y_train.shape)
# outcome만 들어있기에 (x, )=> 형태로 나옴
y_train.head()
```

```python
X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
y_test = test[label_name]
print(y_test.shape)
y_test.head()
```

```python
# 머신러닝 알고리즘 !
from sklearn.tree import DecisionTreeClassifier # 디시전트리는 스무고개 같은 아이

model = DecisionTreeClassifier(random_state=42)
# random state에 따라 값이 달라짐!
```

```python
#학습
model.fit(X_train, y_train)
# 위에서 만든 훈련데이터와 테스트 데이터를 모델에 전달하여 학습시킴
```

```python
#예측
y_predict = model.predict(X_test) 
# 예측은 y의 정답값을 넣지않고 훈련데이터만 넣어줌!
print(y_predict)
y_predict[:5]
```

```python
# 트리 알고리즘 분석
from sklearn.tree import plot_tree # 플롯트리를 가져옴

plot_tree(model, feature_names=feature_names)
# 위에서 사용된 model과 feature_names를 이용함
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
tree = plot_tree(model, feature_names=feature_names,
                           filled=True, 
                           fontsize=10)

# gini가 0이면 맞는 지 아닌 지 확인시켜줌
```

```python
# 실제값 - 예측값을 빼주면 같은 값은 0으로 나오게 되옵니다.
# 여기에서 절대값을 씌운 값이 1인 값이 다르게 예측한 값이 됩니다.
# 44 => 39 => 49(나이 25) => 55(나이 30) => 23 (인슈린 결측치를 평균으로) => 16(인슐린의 결측치를 중앙값으로 대체)
#=> 15(인슐린과 글루코스 상관계수로 파생변수를 생성) => 이상치를 600이하로 하여 설정했을 때 15
# 숫자의 범위가 크다면 scaling 후에 에측의 정확도가 많이 올라갈 수 있음! 
# 가비지인 가비지 아웃이 있음

#데이터가 많으면 결측치를 제거 하는 거보다.. 대체하는 것이 좋을 수 있음!
diff_count = abs(y_test - y_predict).sum()
diff_count
# y.test - y_predict에서 절대값을 한 것이 1이 나온다면 (-1 or 1) 다르게 예측된 값임.. => 그것을 다 더해준 것이 44개라는 것은
# 4x개가 다르게 나왔다는 것임! => x_tsst를 통해 예측한 값과 154개의 테스트 된 값이
# 값이 추가하거나 줄어들 경우에.. 다른것이 줄어들거나 늘어난거임
```

```python
abs(y_test - y_predict).sum() / len(y_test)

(len(y_test) - diff_count) / len(y_test) * 100 # 100점 만점으로 하여 몇개를 정확하게 만들어주었다.
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```

```python
model.score(X_test, y_test) * 100
```
