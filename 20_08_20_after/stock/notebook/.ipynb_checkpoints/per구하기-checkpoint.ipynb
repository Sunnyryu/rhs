```python
import pandas as pd 
corp_info = pd.read_csv('corp.csv', header=0)
corp_info
```

```python
corp_info = corp_info[{'종목코드','종목명','현재가','상장주식수'}]
corp_info
```

```python
corp_info['종목코드'] = 'A' + corp_info['종목코드']
corp_info
```

```python
corp_info['종목코드'].apply(lambda x:'A'+x)
```

```python
def addA(value):
    return 'A'+str(value)

corp_info['종목코드'].apply(addA)
```

```python
def double_data(value):
    return value*2
modified_data = corp_info.apply(double_data)
modified_data
```

```python
import openpyxl
corp_result = pd.read_excel('SAMPLE_2018.xlsx', engine ='openpyxl')
corp_result

```

```python
corp_result.columns

corp_data = corp_result[{'종목코드', '당기순이익'}]
corp_data
```

```python
corp_total = pd.merge(corp_info, corp_data, how='left', on='종목코드')
corp_total
```

```python
corp_total = corp_info.join(corp_data.set_index('종목코드'), how='outer', on='종목코드')
corp_total
```

```python
corp_total["EPS"] = corp_total['당기순이익']/corp_total['상장주식수'].str.replace(',','').astype('float')
corp_total
```

```python
corp_total = corp_total[ pd.notnull(corp_total['EPS'])]
corp_total = corp_total[ pd.notnull(corp_total['현재가'])]

corp_total
```

```python
corp_total["PER"] = corp_total["현재가"].str.replace(",","").astype('int')/corp_total["EPS"]
corp_total
```

```python
corp_filter = corp_total[(corp_total["PER"]>=12) & (corp_total["PER"]<=15)]
corp_filter
```

```python
corp_filter.sort_values(by='PER')
```

```python
corp_filter.sort_values(by='PER', ascending=False)
```

```python
corp_filter.sort_values(by=['PER','EPS'], ascending=False)
```
